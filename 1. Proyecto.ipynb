```python
duplicados = orderProductsTrain.drop_duplicates()
```

```python
duplicados.shape
```

```python
from scipy import stats
import numpy as np

def outliers_col(duplicados):
    for columna in duplicados:
        if duplicados[columna].dtype != np.object:
            n_outliers = len(duplicados[np.abs(stats.zscore(duplicados[columna])) > 3])
            print("{} | {} | {}".format(
                duplicados[columna].name,
                n_outliers,
                duplicados[columna].dtype
            ))
    
outliers_col(duplicados)
```

```python
duplicados.describe()
```

```python
orden_carrito_min = duplicados.orden_agregado_carrito.mean() - duplicados.orden_agregado_carrito.std()
orden_carrito_max = duplicados.orden_agregado_carrito.mean() + 4*duplicados.orden_agregado_carrito.std()

print(orden_carrito_min,orden_carrito_max)
```
